```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples = 200, centers = 6, random_state = 42)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
plt.show()
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X_moons, y_moons = make_moons(n_samples=100, noise=0.1)

plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, s=50, cmap='viridis')
plt.show()
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X_circles, y_circles = make_circles(n_samples=1000, noise=0.1)

plt.scatter(X_circles[:, 0], X_circles[:, 1], c=y_circles, s=50, cmap='viridis')
plt.show()
```

```python
from sklearn.datasets import make_gaussian_quantiles
import matplotlib.pyplot as plt

X, y = make_gaussian_quantiles(n_samples = 300, n_features = 2, n_classes = 4)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.show()
```

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

X, y = make_classification(n_samples = 1000, n_features = 20, n_informative = 2, n_redundant = 10, weights = [0.95, 0.05])


plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
plt.title("Imbalanced Dataset")
plt.show()
```

```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt

X, y = make_regression(n_samples = 100, n_features = 1, noise = 0.3)

deviation = np.random.normal(loc = 0, scale = 10, size = y.shape[0])
y_with_deviation = y + deviation

plt.scatter(X, y_with_deviation, color = 'blue', s = 30, marker = 'o', label = "Data points")
plt.show()
```
